```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# 随机生成100个数据
data = pd.Series(np.random.randn(100),index=np.arange(100))
 
# 为了方便观看效果, 我们累加这个数据
data.cumsum()

# pandas 数据可以直接观看其可视化形式
data.plot()
```
